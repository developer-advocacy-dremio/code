```python
from pyiceberg.table import StaticTable
```

```python
table = StaticTable.from_metadata(
    "s3a://my-bucket/test.db/salesnew/metadata/00001-91773d96-d1f4-490e-a894-9bc0652b6500.metadata.json"
)
```

```python
table
```

```python
scan = table.scan()
```

```python
con = table.scan().to_duckdb(table_name="sales")
```

```python
duck_data = con.execute("SELECT product, price FROM sales GROUP BY ALL")
duck_df = duck_data.df()
duck_df
```
